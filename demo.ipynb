```python
import cv2
from cvzone.HandTrackingModule import HandDetector as hd
```

```python
cap = cv2.VideoCapture(1)
```

```python
detector = hd(detectionCon=int(0.8))
```

```python
startDistance = None
scale = 0
cx, cy = 200, 200 
```

```python
while True:
    succ, img = cap.read()
    result = detector.findHands(img)
    if isinstance(result, tuple) and len(result) == 2:
        hands, img = result
    else:
        hands = result
        img = None

    if img is None:
        print("Image processing failed")
        continue

    img1 = cv2.imread("boy.jpeg")

    if len(hands) == 2:
        hand1 = hands[0]
        hand2 = hands[1]

        hand1_fingers = detector.fingersUp(hand1)
        hand2_fingers = detector.fingersUp(hand2)

        if hand1_fingers == [1, 1, 0, 0, 0] and hand2_fingers == [1, 1, 0, 0, 0]:
            ImList1 = hand1["lmlist"]
            ImList2 = hand2["lmList"]

            if startDistance is None:
                length, info, img = detector.findDistance(hand1["center"], hand2["center"], img)
                scale = int((length - startDistance) // 2)
                cx, cy = info[4:]
    
    else:
        startDistance = None

    try:
        h1, w1, _ = img.shape
        newH, newW = ((h1 + scale) // 2) *2, ((w1 + scale) // 2) *2
        img1 = cv2.resize(img1, (newH, newW))
        img[cy - newH//2 : cy + newH//2, cx - newW//2 : cx + newW//2] = img1

    except:
        pass

    cv2.imshow("SEX", img)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
