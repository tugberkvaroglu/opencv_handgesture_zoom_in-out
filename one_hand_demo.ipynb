```python
import cv2
from cvzone.HandTrackingModule import HandDetector as hd
```

```python
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cant open the video")

cap.set(3, 1280)
cap.set(4, 720)
```

```python
detector = hd(detectionCon=int(0.8))
```

```python
startDistance = None
scale = 0
cx, cy = 200, 200 
```

```python
while True:
    succ, img = cap.read()
    hands, img = detector.findHands(img)
    img1 = cv2.imread("boy.jpeg")

    if len(hands) == 1:

        if detector.fingersUp(hands[0]) == [1, 1, 0, 0, 0]:
            lmList1 = hands[0]["lmList"]

            if startDistance == None:
                length, info, img = detector.findDistance(lmList1[4], lmList1[8] , img)
                print(length)
                startDistance = length
            
            length, info, img = detector.findDistance(lmList1[4], lmList1[8] , img)
            scale = int(length - startDistance) * 2

            cx, cy = info[4:]

            print(scale)
    else:
        startDistance = None

    try:
        h1, w1, _ = img1.shape

        newH, newW = (((h1 + scale) // 2) * 2), (((w1 + scale) // 2) * 2)

        img1 = cv2.resize(img1, (newW, newH))

        img[cy - newH // 2 : cy + newH // 2, cx - newW // 2 : cx + newW // 2] = img1

    except:
        pass    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

    cv2.imshow("Image", img)
    

cap.release()
cv2.destroyAllWindows()
```
